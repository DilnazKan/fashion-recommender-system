# 02 — Preprocessing

Build the synthesized text "content profile" per product and scale numeric features, using `src/preprocess.py`.

```python
import sys
sys.path.append("..")

import pandas as pd
from src.data_loader import load_raw_data
from src.preprocess import build_content_profile

df = load_raw_data("../data/raw/fashion_products.csv")
df.head()
```

## Build content profile + scaled numeric features

```python
processed_df = build_content_profile(df)
processed_df[["Product ID", "content_profile", "Price", "Price_scaled", "Rating", "Rating_scaled"]].head(10)
```

Each product's `content_profile` concatenates its categorical attributes into a short pseudo-description, e.g. `"Dress Adidas Men's Fashion Black XL"`. `Price` and `Rating` are min-max scaled to [0, 1] so they can be blended with the TF-IDF text vectors without one feature type dominating the similarity score.

```python
print("Price scaled range:", processed_df["Price_scaled"].min(), "-", processed_df["Price_scaled"].max())
print("Rating scaled range:", processed_df["Rating_scaled"].min(), "-", processed_df["Rating_scaled"].max())
```

## Save cleaned dataset

```python
processed_df.to_csv("../data/processed/cleaned_fashion_products.csv", index=False)
print("Saved", len(processed_df), "rows to ../data/processed/cleaned_fashion_products.csv")
```
